```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset= pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:, [3,4]].values
```

```python
wcss=[]

for i in range(1,11): 
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10,random_state=0 )
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.xlabel("Clusters' amount")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init ='k-means++', max_iter=300, n_init=10,random_state=0 )
y_kmeans = kmeans.fit_predict(X)
```

I will use 5 clusters

```python
fig, ax = plt.subplots(1, 1,figsize=(20,10))

#printing elements to see, how clusterisation works
ax.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], s=100, c='blue', label ='Cluster 1')
ax.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=100, c='green', label ='Cluster 2')
ax.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=100, c='red', label ='Cluster 3')
ax.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=100, c='magenta', label ='Cluster 4')
ax.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], s=100, c='cyan', label ='Cluster 5')

#printing centers of each cluster
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')

ax.legend()
ax.grid()
ax.set(xlabel='Annual Income', ylabel='Spending Score');
```
